```python
import tensorflow as tf
import numpy as np
```

```python
mobile=tf.keras.applications.mobilenet.MobileNet()
```

```python
from tensorflow.keras.preprocessing import image
def pre_processing_img(img_name):
    img_path='images/'
    img=image.load_img(img_path+img_name,target_size=(224,224))
    img_array=image.img_to_array(img)
    expanded_img=np.expand_dims(img_array,axis=0)
    return tf.keras.applications.mobilenet.preprocess_input(expanded_img)    
```

```python
from IPython.display import Image

Image(filename='images/lemon.jpg',width=300,height=300)
```

```python
pre_processed_img=pre_processing_img('lemon.jpg')
predictions=mobile.predict(pre_processed_img)
print(predictions)
```

```python
from tensorflow.keras.applications import imagenet_utils

results=imagenet_utils.decode_predictions(predictions)
print(results)
```

```python
print(results[0][0]) #best prediction
```

```python
Image(filename='images/pineapple.jpg',width=300,height=300)
```

```python
pre_processed_img=pre_processing_img('pineapple.jpg')
predictions=mobile.predict(pre_processed_img)
results=imagenet_utils.decode_predictions(predictions)
print(results[0][0])
```
